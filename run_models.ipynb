```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv('export.csv')  # your full dataset

# Filter for valid survival and cause info
df = df[
    df['Survival months'].notna() &
    df['SEER cause-specific death classification'].notna() &
    df['COD to site recode'].notna()
]

# Optional: filter by diagnosis year
df = df[df['Year of diagnosis'].astype(int) >= 2000]

# Stratify by outcome
df_sampled, _ = train_test_split(
    df,
    stratify=df['SEER cause-specific death classification'],
    train_size=10000,
    random_state=42
)

df_sampled.to_csv('export_10k.csv', index=False)
```
