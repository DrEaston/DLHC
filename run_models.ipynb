```python
import pandas as pd

file_path = '/content/drive/MyDrive/export.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
df.head(100).to_csv('/content/drive/MyDrive/export_100rows.csv', index=False)
```

```python
# Filter to only breast cancer cases (ICD Site Code ~500–509 usually maps to breast)
df = df[df['Primary Site'].between(500, 509)]
```

```python
# Remove cases without survival time or cause of death info
df = df[df['Survival months'].notna()]
df = df[df['Vital status recode (study cutoff used)'].notna()]
df = df[df['COD to site recode'].notna()]
```

```python
# Randomly sample 10,000 rows
df_sample = df.sample(n=10000, random_state=42)
```

```python
df_sample.to_csv('/content/drive/MyDrive/seer_sample_10k.csv', index=False)
```
